```python
q1 = data['age'].quantile(0.25)
q3 = data['age'].quantile(0.75)
iqr = q3 - q1
lower_bond = q1 - 1.5*iqr
upper_bond = q3 + 1.5*iqr
med = data['age'].median()
data['age'] = data['age'].apply(lambda x: med if x > upper_bond or x < lower_bond else x)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x,y = smote.fit_resample(data.drop('Class/ASD', axis=1), data['Class/ASD'])
print(y.value_counts())
```

```python
#resampled_data1 = pd.concat([pd.DataFrame(x,columns=data.drop('Class/ASD', axis=1).columns), pd.DataFrame(y,columns=['Class/ASD'])], axis=1)
resampled_data = pd.concat([pd.DataFrame(x,columns = data.iloc[:,:-1].columns), pd.DataFrame(y,columns=['Class/ASD'])], axis=1)
```

```python
resampled_data.head()
```

```python
data.head()
```

```python
data = resampled_data
```

```python
data['Class/ASD'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data.iloc[:,:-1], data['Class/ASD'], test_size=0.2, random_state=42)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
#from sklearn.svm import SVC
#from catboost import CatBoostClassifier 
from  xgboost import XGBClassifier
#from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, recall_score,precision_score, f1_score,make_scorer
```

```python
models = {
    'DesicionTree':DecisionTreeClassifier(random_state=42),
    'RandomForest':RandomForestClassifier(random_state=42),
    'XGB':XGBClassifier(random_state=42),
}
```

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    print('----------',name,'----------')
    print('accuracy:',accuracy_score(pred,y_test))
    print('precision:',precision_score(pred,y_test))
    print('recall:',recall_score(pred,y_test))
    print('f1_score:',f1_score(pred,y_test))
    print('confusion_matrix:',confusion_matrix(pred,y_test))
    print('classification_report:',classification_report(pred,y_test))
    print('------------------------------------------------------------------------------')
```

```python
from sklearn.model_selection import cross_val_score,StratifiedKFold
```

```python
scores = {}
kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
for name,model in models.items():
    cv_results = cross_val_score(model, x_train, y_train, cv=kfold, scoring="accuracy")
    scores[name] = cv_results
    print('----------',name,'----------')
    print(f'Accuracy: {np.mean(cv_results):.2f}')
    print('------------------------------------------------------------------------------')
```

```python
param_dt = {
    'criterion':['gini','entropy'],
    'max_depth':[5,10,15,20,25,30],
    'min_samples_split':[2,5,10,15,20],
    'min_samples_leaf':[1,2,5]
}
param_rf = {
    'n_estimators':[100,200,300,500],
    'criterion':['gini','entropy'],
    'max_depth':[5,10,15,20,25,30],
    'min_samples_split':[2,5,10,15,20],
    'min_samples_leaf':[1,2,5],
    'bootstrap':[True,False]
}
param_xgb = {
    'n_estimators':[100,200,300,400,500],
    'max_depth':[5,10,15,20,25,30],
    'learning_rate':[0.05,0.1,0.15,0.2],
    'subsample':[0.5,0.6,0.7,0.8,0.9,1],
    'colsample_bytree':[0.5,0.6,0.7,0.8,0.9,1],
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
xgb_model = RandomizedSearchCV(XGBClassifier(random_state=42), param_xgb, cv=kfold, scoring='accuracy', n_iter=20, random_state=42)
rf_model = RandomizedSearchCV(RandomForestClassifier(random_state=42), param_rf, cv=kfold, scoring='accuracy', n_iter=20, random_state=42)
dt_model = RandomizedSearchCV(DecisionTreeClassifier(random_state=42), param_dt, cv=kfold, scoring='accuracy', n_iter=20, random_state=42)
```

```python
xgb_model.fit(x_train, y_train)
rf_model.fit(x_train, y_train)
dt_model.fit(x_train, y_train)
```

```python
print(xgb_model.best_estimator_)
print(xgb_model.best_params_)
print(xgb_model.best_score_)
print('--------------------------------------------------------')
print(rf_model.best_estimator_)
print(rf_model.best_params_)
print(rf_model.best_score_)
print('--------------------------------------------------------')
print(dt_model.best_estimator_)
print(dt_model.best_params_)
print(dt_model.best_score_)
```

```python
best_model = None
best_score = 0
for model in [xgb_model, rf_model, dt_model]:
    if model.best_score_ > best_score:
        best_score = model.best_score_
        best_model = model
```

```python
pred = best_model.predict(x_test)
print('accuracy:',accuracy_score(pred,y_test))
print('precision:',precision_score(pred,y_test))
print('recall:',recall_score(pred,y_test))
print('f1_score:',f1_score(pred,y_test))
print('confusion_matrix:',confusion_matrix(pred,y_test))
print('classification_report:',classification_report(pred,y_test))
```
